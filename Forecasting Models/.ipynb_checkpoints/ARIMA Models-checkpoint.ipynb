# PHASE 1: Load Data 

```python
df = pd.read_csv(r"DATASETS\ForForecast.csv")
df['TransactionDateTime'] = pd.to_datetime(df['TransactionDateTime'])
df.set_index('TransactionDateTime', inplace=True)
df = df.sort_index()

print(f"Shape {df.shape}")
df.tail(10)
```

```python
df.dtypes
```

# PHASE 2: Prepare Data

## Resample and fix

```python
desired_column = 'NetAmountWOVat'
```

```python
df = df[[desired_column]]
```

```python
resampled_df = df.resample('60T', closed="left").sum()
resampled_df.shape
```

```python
resampled_df.head(10)
```

```python
plt.figure(figsize=(14, 7))

plt.plot(resampled_df.index, resampled_df[f'{desired_column}'], label=f'{desired_column}', color='blue')

plt.xlabel('Date')
plt.ylabel(f'{desired_column}')
plt.title(f'Time Series Plot of {desired_column}')

plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45, ha='right')

for date in np.unique(resampled_df.index.date):
    plt.axvline(x=date, color='gray', linestyle='--', linewidth=0.5)

plt.legend()
plt.tight_layout()
plt.show()
```

## Check Seasonality

ARIMA modelleri için verinin sezonsal doğasını anlama önemli, çünkü daha sonra modelde argüman olarak kullanılıyor.

```python
# Parçaları ayrıştırmak
decomposition = seasonal_decompose(resampled_df)

# Parçaları toplamak
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Orjinal serinin grafiği
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(resampled_df, label='Original Time Series')
plt.legend()

# Trend grafiği
plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend')
plt.legend()

# Sezonsal grafik
plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonality')
plt.legend()

# Kalıntı grafiği
plt.subplot(4, 1, 4)
plt.plot(residual, label='Residuals')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
ax = seasonal.plot(label='Seasonality', color='blue')

min_ = seasonal.idxmin()
max_ = seasonal.idxmax()
min_2 = seasonal[min_:].idxmin()
max_2 = seasonal[max_:].idxmax()

ax.axvline(min_, label='min 1', c='red')
ax.axvline(min_2, label='min 2', c='red', ls=':')
ax.axvline(max_, label='max 1', c='green')
ax.axvline(max_2, label='max 2', c='green', ls=':')
plt.legend(loc='upper right', fontsize='x-small')

start_time = seasonal.index[0]
time_difference_hours = (min_ - start_time).total_seconds() / 3600

print(f'The time difference between the two minimums from the beginning is {time_difference_hours} hours')
```

## Other Features

```python
resampled_df["HourInterval"] = resampled_df.index.hour
resampled_df['DayOfWeek'] = resampled_df.index.dayofweek + 1

resampled_df = resampled_df.dropna()
```

```python
print(resampled_df.shape)
resampled_df.head()
```

```python
features = ['HourInterval', 'DayOfWeek']
```

## Train-Test Split

```python
train_ratio = 0.8
train_size = int(len(resampled_df) * train_ratio)

train_df = resampled_df[:train_size]
valid_df = resampled_df[train_size:]
```

```python
train_df
```

```python
valid_size = len(resampled_df) - train_size
print("Train size:", train_size, "Valid Size:", valid_size)
```

```python
plt.figure(figsize=(14, 7))

plt.plot(train_df.index, train_df[f'{desired_column}'], label='Training Set', color='blue')
plt.plot(valid_df.index, valid_df[f'{desired_column}'], label='Validation Set', color='orange')

plt.xlabel('Date')
plt.ylabel('Your Y-axis Label')
plt.title('Combined Plot of Training and Validation Sets')

plt.gca().xaxis.set_major_locator(mdates.DayLocator())  # Adjust this based on your data frequency
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  # Adjust the date format as needed

plt.xticks(rotation=45, ha='right')

for date in np.unique(resampled_df.index.date):
    plt.axvline(x=date, color='gray', linestyle='--', linewidth=0.5)

plt.legend()
plt.tight_layout()
plt.show()
```

# PHASE 3: Training

Bu kısımda 3 farklı ARIMA modeli eğitiliyor. Sırasıyla:
* ARIMA: Fazladan değişken kullanmadan ARIMA model ailesine ait bir mimari 
* SARIMA: ARIMA'nın sezon etkisini dikkate alan hali 
* SARIMAX: SARIMA'nın fazladan değişkenleri dikkate alan hali 

pmdarima kütüphanesi ARIMA modelleri için en iyi parametreleri otomatik olarak seçen fonksiyonları içeren bir kütüphane, ben de eğitim sırasında bunu kullandım. Çok fazla veri kullanmak isterseniz eğitim süresi oldukça yavaşlayabilir.

### ARIMA

```python
""" 
Burada -ve diğer modeller için olanlarda da- tanımlanan parametreler ne kadar uzun bir arayış yapacağı ile alakalı, tam olarak ne yaptıklarını
bilmek isterseniz kütüphanenin dokümanlarına bakabilirsiniz.
"""

ARIMA_model = pm.auto_arima(train_df[desired_column], start_p=1, start_q=1, max_p=5,
                           max_q=5, start_P=0, start_Q=0, max_P=5,
                           max_Q=5, m=1, max_order=None,
                           trace=True)
```

```python
ARIMA_model.plot_diagnostics(figsize=(15,12))
plt.show()
```

### SARIMA

```python
SARIMA_model = pm.auto_arima(train_df[desired_column], 
                             start_p=1, start_q=1, max_p=5, max_q=5, start_P=0, start_Q=0, max_P=5, max_Q=5, m=24, 
                             max_order=None, trace=True)
```

```python
SARIMA_model.plot_diagnostics(figsize=(15,12))
plt.show()
```

## SARIMAX 

```python
SARIMAX_model = pm.auto_arima(y=train_df[[f'{desired_column}']], X=train_df[features],
                            start_p=1, start_q=1, max_p=5, max_q=5, start_P=0, start_Q=0, max_P=5, max_Q=5, m=24,
                            max_order=None, trace=True)
```

```python
SARIMAX_model.plot_diagnostics(figsize=(15,12))
plt.show()
```

# PHASE 4: Testing

## Plot inference

```python
n_periods = valid_size
index_of_fc = valid_df.index

## ARIMA
ARIMA_fitted, ARIMA_confint = ARIMA_model.predict(n_periods=n_periods, return_conf_int=True)

ARIMA_fitted_series = pd.Series(ARIMA_fitted, index=index_of_fc)
ARIMA_lower_series = pd.Series(ARIMA_confint[:, 0], index=index_of_fc)
ARIMA_upper_series = pd.Series(ARIMA_confint[:, 1], index=index_of_fc)

## SARIMA
SARIMA_fitted, SARIMA_confint = SARIMA_model.predict(n_periods=n_periods, return_conf_int=True)

SARIMA_fitted_series = pd.Series(SARIMA_fitted, index=index_of_fc)
SARIMA_lower_series = pd.Series(SARIMA_confint[:, 0], index=index_of_fc)
SARIMA_upper_series = pd.Series(SARIMA_confint[:, 1], index=index_of_fc)

## SARIMAX
SARIMAX_fitted, SARIMAX_confint = SARIMAX_model.predict(X=valid_df[features], n_periods=n_periods, return_conf_int=True)

SARIMAX_fitted_series = pd.Series(SARIMAX_fitted, index=index_of_fc)
SARIMAX_lower_series = pd.Series(SARIMAX_confint[:, 0], index=index_of_fc)
SARIMAX_upper_series = pd.Series(SARIMAX_confint[:, 1], index=index_of_fc)

# Ensemble Predictions
ensemble_predictions = (ARIMA_fitted + SARIMA_fitted + SARIMAX_fitted) / 3
ensemble_lower_series = (ARIMA_lower_series + SARIMA_lower_series + SARIMAX_lower_series) / 3
ensemble_upper_series = (ARIMA_upper_series + SARIMA_upper_series + SARIMAX_upper_series) / 3
```

```python
# Plot
plt.figure(figsize=(15,7))
plt.plot(train_df[desired_column], color='#1f76b4')
plt.plot(ARIMA_fitted_series, color='darkgreen')
plt.fill_between(ARIMA_lower_series.index, ARIMA_lower_series, ARIMA_upper_series, color='k', alpha=.15)

plt.title("ARIMA Forecast")
plt.show()

# Plot
plt.figure(figsize=(15,7))
plt.plot(train_df[desired_column], color='#1f76b4')
plt.plot(SARIMA_fitted_series, color='darkgreen')
plt.fill_between(SARIMA_lower_series.index, SARIMA_lower_series, SARIMA_upper_series, color='k', alpha=.15)

plt.title("SARIMA Forecast")
plt.show()

# Plot
plt.figure(figsize=(15,7))
plt.plot(train_df[desired_column], color='#1f76b4')
plt.plot(SARIMAX_fitted_series, color='darkgreen')
plt.fill_between(SARIMAX_lower_series.index, SARIMAX_lower_series, SARIMAX_upper_series, color='k', alpha=.15)

plt.title("SARIMAX Forecast")
plt.show()

# Plot
plt.figure(figsize=(15,7))
plt.plot(train_df[desired_column], color='#1f76b4')
plt.plot(ensemble_predictions, color='darkgreen')
plt.fill_between(SARIMAX_lower_series.index, ensemble_lower_series, ensemble_upper_series, color='k', alpha=.15)

plt.title("Ensemble Forecast")
plt.show()
```

```python
result_df = valid_df.merge(pd.DataFrame({f'{desired_column} ARIMA': ARIMA_fitted, f'{desired_column} SARIMA': SARIMA_fitted, f'{desired_column} SARIMAX': SARIMAX_fitted, f'{desired_column} Ensemble': ensemble_predictions}, index=valid_df.index), left_index=True, right_index=True)
result_df.tail(10)
```

## Table Inference

```python
result_df["Original_Status"] = np.where(result_df[f"{desired_column}"].shift(-1) > result_df[f"{desired_column}"], "Will Rise", "Will Fall")
result_df["Predicted_Status_ARIMA"] = np.where(result_df[f"{desired_column} ARIMA"].shift(-1) > result_df[f"{desired_column} ARIMA"], "Will Rise", "Will Fall")
result_df["Predicted_Status_SARIMA"] = np.where(result_df[f"{desired_column} SARIMA"].shift(-1) > result_df[f"{desired_column} SARIMA"], "Will Rise", "Will Fall")
result_df["Predicted_Status_SARIMAX"] = np.where(result_df[f"{desired_column} SARIMAX"].shift(-1) > result_df[f"{desired_column} SARIMAX"], "Will Rise", "Will Fall")
result_df["Predicted_Status_Ensemble"] = np.where(result_df[f"{desired_column} Ensemble"].shift(-1) > result_df[f"{desired_column} Ensemble"], "Will Rise", "Will Fall")

result_df.head(1000)
```

```python
real_values = result_df[f'{desired_column}'].sum()
ARIMA_values = result_df[f'{desired_column} ARIMA'].sum()
SARIMA_values = result_df[f'{desired_column} SARIMA'].sum()
SARIMAX_values = result_df[f'{desired_column} SARIMAX'].sum()
ensemble_values = result_df[f'{desired_column} Ensemble'].sum()

print(f"Real {desired_column}:", real_values)
print(f"ARIMA {desired_column}:", ARIMA_values)
print(f"SARIMA {desired_column}:", SARIMA_values)
print(f"SARIMAX {desired_column}:", SARIMAX_values)
print(f"Ensemble {desired_column}:", ensemble_values)

print("-------------------------------------------")

print("The prediction offness:")
print("ARIMA:", int(ARIMA_values - real_values))
print("SARIMA:", int(SARIMA_values - real_values))
print("SARIMAX:", int(SARIMAX_values - real_values))
print("Ensemble:", int(ensemble_values - real_values))
```

```python
distance_ARIMA = ARIMA_values - real_values
distance_SARIMA = SARIMA_values - real_values
distance_SARIMAX = SARIMAX_values - real_values
distance_ensemble = ensemble_values - real_values

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

axes[0].bar(['Real', 'ARIMA', 'SARIMA', 'SARIMAX', 'Ensemble'],
            [real_values, ARIMA_values, SARIMA_values, SARIMAX_values, ensemble_values],
            color=['blue', 'orange', 'green', 'red', 'yellow'])
axes[0].set_title('Total Values for All Models')
axes[0].set_ylabel(f'Total {desired_column}')

axes[1].bar(['ARIMA', 'SARIMA', 'SARIMAX', 'Ensemble'],
            [distance_ARIMA, distance_SARIMA, distance_SARIMAX, distance_ensemble],
            color=['orange', 'green', 'red', 'yellow'])
axes[1].set_title(f'Distance to Real {desired_column}')
axes[1].set_ylabel('Absolute Distance')

plt.tight_layout()

plt.show()
```

# PHASE 5: Forecasting

## Prepare the emtpy set

```python
start_timestamp = '2024-02-09 00:00:00'
end_timestamp = '2024-02-18 23:00:00'

forecast_timestamps = pd.date_range(start=start_timestamp, end=end_timestamp, freq='H')

forecast_df = pd.DataFrame(index=forecast_timestamps)

forecast_df[f'Forecasted {desired_column}'] = np.nan

forecast_df["HourInterval"] = forecast_df.index.hour
forecast_df['DayOfWeek'] = forecast_df.index.dayofweek + 1

forecast_df = forecast_df[[f'Forecasted {desired_column}', 'HourInterval', 'DayOfWeek']]

forecast_df.head(4)
```

```python
print(forecast_df.shape)
forecast_df.head(4)
```

## Forecasts

```python
ARIMA_forecast_df = forecast_df.copy()
SARIMA_forecast_df = forecast_df.copy()
SARIMAX_forecast_df = forecast_df.copy()
```

```python
n_periods = len(forecast_df)
index_of_fc = forecast_df.index

########################## ARIMA ##################################
ARIMA_model = pm.auto_arima(resampled_df[desired_column], start_p=1, start_q=1, max_p=5,
                           max_q=5, start_P=0, start_Q=0, max_P=5,
                           max_Q=5, m=1, max_order=None,
                           trace=False)

## ARIMA
ARIMA_fitted, ARIMA_confint = ARIMA_model.predict(n_periods=n_periods, return_conf_int=True)

ARIMA_fitted_series = pd.Series(ARIMA_fitted, index=index_of_fc)
ARIMA_lower_series = pd.Series(ARIMA_confint[:, 0], index=index_of_fc)
ARIMA_upper_series = pd.Series(ARIMA_confint[:, 1], index=index_of_fc)

########################## SARIMA ##################################
SARIMA_model = pm.auto_arima(resampled_df[desired_column], start_p=1, start_q=1, max_p=5,
                           max_q=5, start_P=0, start_Q=0, max_P=5,
                           max_Q=5, m=24, max_order=None,
                           trace=False)

## SARIMA
SARIMA_fitted, SARIMA_confint = SARIMA_model.predict(n_periods=n_periods, return_conf_int=True)

SARIMA_fitted_series = pd.Series(SARIMA_fitted, index=index_of_fc)
SARIMA_lower_series = pd.Series(SARIMA_confint[:, 0], index=index_of_fc)
SARIMA_upper_series = pd.Series(SARIMA_confint[:, 1], index=index_of_fc)

########################## SARIMA ##################################
SARIMAX_model = pm.auto_arima(y=resampled_df[[f'{desired_column}']], X=resampled_df[features],
                            start_p=1, start_q=1, max_p=5, max_q=5, start_P=0, start_Q=0, max_P=5, max_Q=5, m=24,
                            max_order=None, trace=False)

## SARIMAX
SARIMAX_fitted, SARIMAX_confint = SARIMAX_model.predict(X=forecast_df[features], n_periods=n_periods, return_conf_int=True)

SARIMAX_fitted_series = pd.Series(SARIMAX_fitted, index=index_of_fc)
SARIMAX_lower_series = pd.Series(SARIMAX_confint[:, 0], index=index_of_fc)
SARIMAX_upper_series = pd.Series(SARIMAX_confint[:, 1], index=index_of_fc)
```

```python
# Ensemble
ensemble_predictions = (ARIMA_fitted + SARIMA_fitted + SARIMAX_fitted) / 3
ensemble_lower_series = (ARIMA_lower_series + SARIMA_lower_series + SARIMAX_lower_series) / 3
ensemble_upper_series = (ARIMA_upper_series + SARIMA_upper_series + SARIMAX_upper_series) / 3
```

## Plot Inference

```python
plt.figure(figsize=(15,7))

plt.plot(ARIMA_fitted_series, color='darkgreen')
plt.fill_between(ARIMA_lower_series.index, ARIMA_lower_series, ARIMA_upper_series, color='k', alpha=.15)

plt.title("ARIMA Forecast")
plt.show()

plt.figure(figsize=(15,7))

plt.plot(SARIMA_fitted_series, color='darkgreen')
plt.fill_between(SARIMA_lower_series.index, SARIMA_lower_series, SARIMA_upper_series, color='k', alpha=.15)

plt.title("SARIMA Forecast")
plt.show()

plt.figure(figsize=(15,7))

plt.plot(SARIMAX_fitted_series, color='darkgreen')
plt.fill_between(SARIMAX_lower_series.index, SARIMAX_lower_series, SARIMAX_upper_series, color='k', alpha=.15)

plt.title("SARIMAX Forecast")
plt.show()

plt.figure(figsize=(15,7))
plt.plot(ensemble_predictions, color='darkgreen')
plt.fill_between(SARIMAX_lower_series.index, ensemble_lower_series, ensemble_upper_series, color='k', alpha=.15)

plt.title("Ensemble Forecast")
plt.show()
```
