# Process Categorical Variables

Bu kısımda eşşiz değer sayısı bir hayli fazla olan veriyi ileride daha rahat kullanılabilmesi için basitleştirme işlemleri yer alıyor.

```python
df_forecast.DataAreaKey.value_counts()
```

```python
df_forecast.RboStoreKey.value_counts()
```

```python
df_forecast.CampaignKrcKey.value_counts()
```

```python
""" Burada bazı kategorik sütunlar için frekansı belli bir limitin üzerinde olmayan eşşiz verileri basitlik adına "others" ile değiştirdim."""

# Invent Key
threshold = 50
product_counts = df_forecast['InventKey'].value_counts()
products_to_keep = product_counts[product_counts > threshold].index
df_forecast['InventKey'] = df_forecast['InventKey'].where(df_forecast['InventKey'].isin(products_to_keep), 'others')

# Invent Location Key
threshold = 100
location_counts = df_forecast['InventLocationKey'].value_counts()
locations_to_keep = location_counts[location_counts > threshold].index
df_forecast['InventLocationKey'] = df_forecast['InventLocationKey'].where(df_forecast['InventLocationKey'].isin(locations_to_keep), 'others')

# Campaign Key
threshold = 50
campaign_counts = df_forecast['CampaignKrcKey'].value_counts()
campaigns_to_keep = campaign_counts[campaign_counts > threshold].index
df_forecast['CampaignKrcKey'] = df_forecast['CampaignKrcKey'].where(df_forecast['CampaignKrcKey'].isin(campaigns_to_keep), 'others')

# Rbo Store Key
threshold = 100
store_counts = df_forecast['RboStoreKey'].value_counts()
stores_to_keep = store_counts[store_counts > threshold].index
df_forecast['RboStoreKey'] = df_forecast['RboStoreKey'].where(df_forecast['RboStoreKey'].isin(stores_to_keep), 'others')
```

```python
print(df_forecast.InventKey.value_counts())
print(df_forecast.InventLocationKey.value_counts())
print(df_forecast.CampaignKrcKey.value_counts())
print(df_forecast.RboStoreKey.value_counts())
```
